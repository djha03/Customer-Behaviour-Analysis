```python
import pandas as pd

df= pd.read_csv('customer_shopping_behavior.csv')
```

```python
# Fill missing values in Review Rating using the median rating of each Category.

df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#converted all the columns to lower case, after that replcaed all the spaces with '_' in column name and lastly because purchase amount column lenth was big so did dhorten.

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace("  ","_")
df= df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
# Segregation of columns in age group
df[['age', 'age_group']].head(5)
```

```python
#create column purchase_frequency_days


frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
